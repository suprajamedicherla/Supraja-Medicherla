**Transforming Data Types**

```python
countyweather['min_temperature_air_2m_f']=countyweather['min_temperature_air_2m_f'].astype('int64')
countyweather['avg_temperature_air_2m_f']=countyweather['avg_temperature_air_2m_f'].astype('int64')
countyweather['max_temperature_air_2m_f']=countyweather['max_temperature_air_2m_f'].astype('int64')
countyweather['min_humidity_specific_2m_gpkg']=countyweather['min_humidity_specific_2m_gpkg'].astype('int64')
countyweather['avg_humidity_specific_2m_gpkg']=countyweather['avg_humidity_specific_2m_gpkg'].astype('int64')
countyweather['max_humidity_specific_2m_gpkg']=countyweather['max_humidity_specific_2m_gpkg'].astype('int64')
```

```python
countyweather.info()
```

```python
countyweather.shape
```

```python
countyweather.head()
```

**Dropping Unneeded Columns**

```python
countyweather = countyweather.drop(['country', 'date'], axis=1)
```

```python
countyweather
```

```python
countyweather.info()
```

```python
countyweather.describe()
```

**Checking Null values**

```python
countyweather.isnull()
```

```python
countyweather.dropna(inplace=True)
countyweather
```

**Renaming Columns**

```python
countyweather = countyweather.rename(columns={'county_fips_code': 'County fips','min_temperature_air_2m_f':'Min temp','avg_temperature_air_2m_f':'Avg temp',
                                              'max_temperature_air_2m_f':'Max temp','min_humidity_relative_2m_pct':'Min Relative Humidity','avg_humidity_relative_2m_pct':'Avg Relative Humidity','max_humidity_relative_2m_pct':'Max Relative Humidity',
                             'min_humidity_specific_2m_gpkg':'Min Specific Humidity','avg_humidity_specific_2m_gpkg':'Avg Specific Humidity','max_humidity_specific_2m_gpkg':'Max Specific Humidity'})
```

```python
countyweather
```

**Creating a calculated field based on two other columns**

```python
countyweather['Specific Relative_Avg'] = countyweather['Avg Relative Humidity']+ countyweather['Avg Specific Humidity']
```

```python
countyweather
```

**Removing Duplicate rows**

```python
duplicates = countyweather.duplicated()
print(duplicates)
```

```python
countyweather = countyweather.drop_duplicates()

countyweather
```

**Outlier Analysis**

```python
upper_limit = countyweather['Min temp'].mean() + 3*countyweather['Min temp'].std()
lower_limit = countyweather['Min temp'].mean() - 3*countyweather['Min temp'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
countyweather.loc[(countyweather['Min temp'] > upper_limit) & (countyweather['Min temp'] < lower_limit)]
```

```python
upper_limit = countyweather['Avg temp'].mean() + 3*countyweather['Avg temp'].std()
lower_limit = countyweather['Avg temp'].mean() - 3*countyweather['Avg temp'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
countyweather.loc[(countyweather['Avg temp'] > upper_limit) | (countyweather['Avg temp'] < lower_limit)]
```

```python
sns.boxplot(countyweather['Avg temp'])
```

```python
newcounty = countyweather.loc[(countyweather['Avg temp'] < upper_limit) & (countyweather['Avg temp'] > lower_limit)]
print('old data:', len(countyweather))
print('new data:', len(newcounty))
print('outliers:', len(countyweather)-len(newcounty))
```

```python
sns.boxplot(newcounty['Avg temp'])
```
